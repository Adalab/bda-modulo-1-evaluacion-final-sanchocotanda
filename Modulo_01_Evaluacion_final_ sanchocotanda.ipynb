* Agrega un producto al inventario o actualiza su cantidad si ya existe. 
Debe recibir el nombre, precio y cantidad del producto como parámetros.
* Itera a través del inventario y compara los nombres de los productos con el nombre proporcionado.
* Si el producto ya existe, actualiza la cantidad. Si no existe, agrega un nuevo producto al inventario.

```python
def agregar_producto(nombre, precio, cantidad):
    for producto in inventario:
        if producto["nombre"] == nombre:  
            producto["cantidad"] += cantidad  
            print("Cantidad actualizada en nuestro inventario")
            return "Cantidad actualizada" 


    inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad}) 
    print("No teniamos este producto. Queda añadido a nuestro inventario")
```

```python
def ver_inventario(inventario, nombre=None):
    print("Inventario:")
    encontrado = False 
    for producto in inventario:
        if nombre is None:  
            print(f"Nombre: {producto['nombre']}, Precio: {producto['precio']}, Cantidad: {producto['cantidad']}")
        elif producto["nombre"].lower() == nombre.lower():
            print(f"Nombre: {producto['nombre']}, Precio: {producto['precio']}, Cantidad: {producto['cantidad']}")
            encontrado = True
            break
    if nombre is not None and not encontrado:
        print("Ese producto no está en el inventario")
```

2.4 actualizar_stock(nombre, cantidad)

* Actualiza el stock de un producto en el inventario. 
Debe recibir el nombre del producto y la cantidad a agregar o quitar como parámetros. 

Utiliza un bucle for para recorrer el inventario. 

* Busca el producto por nombre.

* Actualiza la cantidad según la entrada del usuario. 
Si el producto no esta en el inventario muestra un mensaje indicándolo.

```python
def actualizar_stock():
    nombre = input("¿Qué producto quieres modificar? ")
    cantidad = int(input("¿Cuánta cantidad quieres agregar (positiva) o quitar (negativa)? "))

    for producto in inventario:
        if producto["nombre"].lower() == nombre.lower():
            producto["cantidad"] += cantidad
            print(f"Stock actualizado. Nuevo stock de {nombre}: {producto['cantidad']}")
            return

    print("Ese producto no está en el inventario")
```

Busca el producto por nombre.
Elimina el producto del inventario si existe.
Si el producto no esta en el inventario muestra un mensaje indicándolo.

```python
def eliminar_producto(nombre):
    for producto in inventario:
        if producto["nombre"].lower() == nombre.lower(): 
            inventario.remove(producto)  
            print("Este artículo se ha eliminado de nuestro inventario")
            return

    else:
        print("Producto no encontrado") 
        
```

2. 6. calcular_valor_inventario()
: Calcula y muestra el valor total del inventario
Utiliza un bucle for para calcular el valor total del inventario.

Itera a través del inventario y suma el valor de cada producto (precio x cantidad). Es decir,
calcula el valor total del inventario. Ejemplo:
- si tenemos 5 camisas que valen 5 euros y 10 calcetines que valen 1 euro, este método tendria que devovler: 35 euros

valor_camisas = 5 * 5 
valor_calcetines = 10 * 1 
valor_camisas + valor_calcetines = 35

```python
def calcular_valor_inventario(inventario, productos=None):
    total = 0

    if productos is None:
        for item in inventario:
            total += item["precio"] * item["cantidad"]
        return total

    productos = [producto.lower() for producto in productos]  
    for item in inventario:
        if item["nombre"].lower() in productos:
            total += item["precio"] * item["cantidad"]

    return total
```

realizar_compra()
: Permite a un cliente realizar una compra seleccionando productos del
inventario. Debe interactuar con el cliente para seleccionar productos y calcular el costo total de la
compra.
Utiliza un bucle 
while
 para permitir al cliente realizar múltiples compras.
Muestra el inventario y solicita al cliente ingresar el nombre del producto que desea comprar.
Registra los productos seleccionados en un carrito y actualiza el inventario.
Calcula el costo total de la compra.

```python
def realizar_compra(inventario):
    carrito = []

    while True:
        busqueda_usuario = input("¿Qué te gustaría comprar?")

        producto_encontrado = None
        for producto in inventario:
            if producto["nombre"].lower() == busqueda_usuario.lower():
                producto_encontrado = producto
                break
       
        if producto_encontrado is None:
            """print("Este producto no está en el inventario")"""
            seguir = input("Lo siento, no tenemos este producto. ¿Quieres seguir comprando? (si/no)")
            if seguir.lower() != "si":
                break
            else:
                continue   

        if producto_encontrado["cantidad"] == 0:
            print(f"Lo siento, pero {producto_encontrado['nombre']} está agotado")
            seguir = input("Lo siento, pero {producto_encontrado['nombre']} está agotado. ¿Quieres seguir comprando? (si/no)")
            if seguir.lower() != "si":
                break
            else:
                continue  
    
        
        cantidad_producto = int(input(f"¿Cuántas unidades de {busqueda_usuario} quieres?"))

      
        if producto_encontrado["cantidad"] >= cantidad_producto:
            producto_encontrado["cantidad"] -= cantidad_producto
            carrito.append({"nombre": busqueda_usuario,"cantidad": cantidad_producto,"precio": producto_encontrado["precio"]})
            print(f"{cantidad_producto} unidades de {busqueda_usuario} añadidas al carrito")

        else:
            
            stock_disponible = producto_encontrado["cantidad"]
            print(f"Solo quedan {stock_disponible} unidades de {busqueda_usuario} ")
            decision = input(f"¿Quieres añadir esas {stock_disponible} unidades? (si/no)")
            if decision.lower() == "si":
                carrito.append({"nombre": busqueda_usuario, "cantidad": stock_disponible, "precio": producto_encontrado["precio"]})
                print(f"{stock_disponible} unidades de {busqueda_usuario} añadidas al carrito")
                producto_encontrado["cantidad"] = 0

        
        seguir = input("¿Quieres seguir comprando? (si/no) ")
        if seguir.lower() != "si":
            break

    
    print("\n Carrito final:")
    total = 0
    for item in carrito:
        importe = item["cantidad"] * item["precio"]
        total += importe
        print(f"{item['cantidad']} x {item['nombre']} = {importe} €")
    print(f"Total a pagar: {total} €")

    return
```
